# Sampling Frequency and the Nyquist Frequency

Key ideas:

* This notebook will plot some digital signals, both below and above the Nyquist frequency
* Signals with frequencies above the Nyquist frequency disappear or are corrupted, because there are not enough samples per cycle

### Create a time array

```python
# -- Import some python libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format='retina'

# -- Define the sampling frequency, fs = 20 Hz
fs = 20

# -- Define the time samples for 1 second of data
# -- The "arange" method creates an array of numbers
time = np.arange(0,1,1/fs)

# -- Show the time array
print(time)
```

### 2 Hz Signal

```python
# -- Try making a 2 Hz sine wave and plotting it
f = 2   # -- A 2 Hz signal
sin2 = np.sin(2*np.pi*f*time)

# -- Plot the sine wave
plt.plot(time, sin2, label='2 Hz')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.legend()
```

## 5 Hz Signal

What happens if we try to plot a signal at the Nyquist frequency?

```python
f = 5  # A 5 Hz signal
sin5 = np.sin(2*np.pi*f*time)

plt.plot(time, sin5, label='5 Hz')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.ylim(-1,1)
plt.legend()
```

## 10 Hz signal (Signal at the Nyquist Frequency)

```python
# -- Try plotting a signal at the Nyquist frequency
# -- Hint: This can't work, because information above the Nyquist frequency is lost!

f = 10
sin10 = np.sin(2*np.pi*f*time)

plt.plot(time, sin10)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.ylim(-1,1)
```

```python
# -- Finally, to see what's going on, let's plot the 10 Hz signal,
# -- both with low and high sample rates

fs_better = 1000
time_better = np.arange(0,1, 1/fs_better)

sin10_better = np.sin(2*np.pi*f*time_better)

plt.plot(time_better, sin10_better, 'grey', label='High sample rate')
plt.plot(time, sin10, '.', label='Low sample rate')
plt.xlabel('time(s)')
plt.ylabel('Amplitude')
plt.legend(loc=1)
```
